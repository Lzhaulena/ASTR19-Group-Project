```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=123)
samples = rng.gamma(shape=2.0, scale=1.5, size=1000)
```

```python
plt.figure(figsize=(8,5))
plt.hist(samples, bins=100, color='C0', edgecolor='k', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Counts')
plt.title('Histogram of 1000 samples from Gamma(shape=2.0, scale=1.5)')
plt.tight_layout()
plt.savefig
```

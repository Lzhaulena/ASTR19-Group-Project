```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def decaying_exp(x, a, b):
    return a * np.exp(-b * x)

def power_law(x, a, b):
    return a / (x**b)

# data
x = np.linspace(0.1, 2*np.pi, 100)  
a_true = 2.0
b_true = 0.5
y_true = decaying_exp(x, a_true, b_true)

sigma = np.linspace(2.0, 0.2, x.size)
rng = np.random.default_rng(seed=42)
noise = rng.normal(loc=0.0, scale=sigma)
y_noisy = y_true + noise
```

```python
def exp_model(x, a, b):
    return a * np.exp(-b * x)

def inv_power_model(x, a, b):
    return a / (x**b)
```

```python
p0_exp = [1.0, 0.1]
popt_exp, pcov_exp = curve_fit(exp_model, x, y_noisy, p0=p0_exp, sigma=sigma, maxfev=10000)
perr_exp = np.sqrt(np.diag(pcov_exp))
p0_pow = [1.0, 0.5]
popt_pow, pcov_pow = curve_fit(inv_power_model, x, y_noisy, p0=p0_pow, sigma=sigma, bounds=([0, -5], [10, 5]), maxfev=10000)
perr_pow = np.sqrt(np.diag(pcov_pow))
```

```python
print(f"Exponential fit: a = {popt_exp[0]:.4f} ± {perr_exp[0]:.4f}, b = {popt_exp[1]:.4f} ± {perr_exp[1]:.4f}")
print(f"Inverse-power fit: a = {popt_pow[0]:.4f} ± {perr_pow[0]:.4f}, b = {popt_pow[1]:.4f} ± {perr_pow[1]:.4f}")
```

```python
x_plot = np.linspace(x.min(), x.max(), 400)
y_exp_fit = exp_model(x_plot, *popt_exp)
y_pow_fit = inv_power_model(x_plot, *popt_pow)

plt.figure(figsize=(8,5))
plt.errorbar(x, y_noisy, yerr=sigma, fmt='o', color='C1', ecolor='gray', elinewidth=1, capsize=3, label='noisy data')
plt.plot(x_plot, y_exp_fit, color='C0', linestyle='-', linewidth=2, label=f'exp fit: a={popt_exp[0]:.3f}, b={popt_exp[1]:.3f}')
plt.plot(x_plot, y_pow_fit, color='C2', linestyle='--', linewidth=2, label=f'power fit: a={popt_pow[0]:.3f}, b={popt_pow[1]:.3f}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data with Exponential and Inverse-Power Best Fits')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('fits_comparison.pdf')
plt.show()
```
